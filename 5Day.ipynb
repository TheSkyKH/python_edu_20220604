```python
rainfall = pd.read_csv("https://raw.githubusercontent.com/SoongMoo/soldesk220604/main/%EC%B0%B8%EA%B3%A0%EC%9E%90%EB%A3%8C/NumPy/data/Seattle2014.csv")
rainfall
rainfall['PRCP']
inches = np.array(rainfall['PRCP']) / 254
inches
print(False+False) # False : 0 , True : 1
print(True+False)
print(True+True)
# 강수량따른 일수
print(np.sum(inches > 0.5)) 
# 강수량이 0.5에서 0.8 사이인 날은 몇일이나 되나요?
print(np.sum((inches > 0.5) & (inches < 0.8))) 
# 강수량이 0.5보다 적게 오고 0.8보다 많이 온날은?
print(np.sum(~((inches > 0.5) & (inches < 0.8)))) 
print(np.sum((inches <= 0.5) | (inches >= 0.8)))
# 비가 안온 날은 몇일 일까요?
print(np.sum(inches == 0)) 
# 비가 온날은 몇일 입니까?
print(np.sum(inches != 0))
```
